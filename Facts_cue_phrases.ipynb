```python
# import required module
import os
import xml.etree.cElementTree as ET;
import re
# assign directory
directory = 'Civil Appeal'
f=open("facts_cue_phrases.txt",'r')
facts=f.readlines()
for i in range(len(facts)):
    facts[i]=facts[i].strip()
    facts[i]=facts[i].lower()
f=open("arguments_cue_phrases.txt",'r')
arguments=f.readlines()
for i in range(len(arguments)):
    arguments[i]=arguments[i].strip()
    arguments[i]=arguments[i].lower()


# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    f1=open(f,'r')
    tree=ET.ElementTree(file=f)
    root=tree.getroot()
    # print(root)
    allpara=[]
    for chld in root:
        # print(chld)
        if(chld.tag=='JudgmentText'):
            # print(chld)
            for para in chld:
                if(para.tag=='P'):
                    txt=para.text
                    if(txt!=None and len(txt)>0):
                        txt=txt.strip()
                        txt=txt.lower()
                        allpara.append(txt)
                # else:
                #     allpara.append("STOP")
    # print(allpara)
    fact_on=0
    data=[]
    for i in allpara:
        if(fact_on==1):
            if(i=='STOP'):
                fact_on=0
            for j in arguments:
                if(re.search(j,i)!=None):
                    fact_on=-1
                    break
        elif(fact_on==0):
            for j in facts:
                if(re.search(j,i)!=None):
                    fact_on=1
        
        if(fact_on==1):
            data.append(i)
    f=os.path.join("result", os.path.splitext(filename)[0]+".txt")
    f1=open(f,"w")
    l=""
    for i in data:
        i=i+'\n'
        l=l+i
    f1.write(l)

    
    # break

                    


    

    
```
